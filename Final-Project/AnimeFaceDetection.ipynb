## 图像处理-检测人脸

```python
def detect(category, filename, output_name,cascade_file = r"D:\lbpcascade_animeface-master\lbpcascade_animeface.xml"):
    if not os.path.isfile(cascade_file):
        raise RuntimeError("%s: not found" % cascade_file)

    cascade = cv2.CascadeClassifier(cascade_file)
    image = cv2.imread(filename, cv2.IMREAD_COLOR)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray = cv2.equalizeHist(gray)
    
    faces = cascade.detectMultiScale(gray,
                                     # detector options
                                     scaleFactor = 1.2,
                                     minNeighbors = 5,
                                     minSize = (24, 24))
    for (x, y, w, h) in faces:
        image = image[y:y+h,x:x+w]
        if image.size != 0:
            #转换图像大小，默认双线性插值
            image = cv2.resize(image,(156,156))
            out_path = r"D://jupyter-notebook//ML-Basic-to-improving//detect_faces//"+str(category)+"//"
            # 不存在文件夹则创建
            if not os.path.exists(out_path):
                os.mkdir(out_path)
            if not os.path.exists(out_path + output_name):
                cv2.imwrite(out_path + output_name, image)
```

## 图像处理-随机裁剪旋转，增加样本

```python
import numpy as np
from PIL import Image
import torchvision.transforms as transforms
```

```python
#随机旋转+中心裁剪
def random_rotation(path,category):
    img = Image.open(path)
    pic_name = path.split('\\')[-1]
    RR = transforms.RandomRotation(degrees=(10, 80))   #degrees为随机旋转的角度
    rr_image = RR(img)
    CenterCrop = transforms.CenterCrop((128,128))   #中心裁剪
    cropped_image = CenterCrop(rr_image)  #PIL.Image.Image
    cropped_image.save('..//detect_faces//'+str(category)+'//rotation'+str(pic_name))
```

```python
#图像随机剪裁和中心剪裁
def crop(path,category):
    img = Image.open(path)
    pic_name = path.split('\\')[-1]
    CenterCrop = transforms.CenterCrop((128,128))   #中心裁剪
    cropped_image = CenterCrop(img)  #PIL.Image.Image
    cropped_image.save('..//detect_faces//'+str(category)+"//cropm"+str(pic_name))
                       
    RandomCrop = transforms.RandomCrop(size=(128, 128))  #随机剪裁
    random_image = RandomCrop(img)
    random_image.save('..//detect_faces//'+str(category)+"//cropr"+str(pic_name))
```

```python
#只进行中心剪裁
def crop_mid(path,category):
    img = Image.open(path)
    pic_name = path.split('\\')[-1]
    CenterCrop = transforms.CenterCrop((128,128))   #中心裁剪
    cropped_image = CenterCrop(img)  #PIL.Image.Image
    cropped_image.save('..//detect_faces//'+str(category)+"//cropm"+str(pic_name))
```

```python
#翻转+中心裁剪
def horizontal_flip(path,category):
    img = Image.open(path)
    pic_name = path.split('\\')[-1]
    HF = transforms.RandomHorizontalFlip(p=1.0)  #p为概率，缺省时默认0.5
    hf_image = HF(img)
    CenterCrop = transforms.CenterCrop((128,128))   #中心裁剪
    cropped_image = CenterCrop(hf_image)  #PIL.Image.Image
    cropped_image.save('..//detect_faces//'+str(category)+"//flip"+str(pic_name))
```

```python
category = 'angry'
for root,dirs,files in os.walk("..//detect_faces//"+str(category)): 
    size = len(files)
    i = 1
    for file in files: 
        path = os.path.join(root,file)
        print("[ %d / %d]" % (i,size))
        i+=1
        crop_mid(path,category)
#         random_rotation(path,category)
#         crop(path,category)
#         horizontal_flip(path,category)
        os.remove(path)
```

## 数据集划分标注

```python
categories = ['disgust','fear','happy','surprise','sad','angry','neutral']
for category in categories:
    for root,dirs,files in os.walk("D://jupyter-notebook//ML-Basic-to-improving//detect_faces//"+str(category)): 
        size = len(files)
        train_size = 0.8 * size
        test_size = 0.2 * size
        train_count = 0
        test_count = 0
        for file in files: 
            file_type = file.split('.')[1]
            if train_count <= train_size:
                new_file_name = str(category) + '_train_' + str(train_count) +'_.'+str(file_type)
                old_path = os.path.join(root,file)
                new_path = os.path.join(root,new_file_name)
                if not os.path.exists(new_path):
                    os.rename(old_path,new_path)
                train_count+=1
            else:
                new_file_name = str(category) + '_test_' + str(test_count) +'_.'+str(file_type)
                old_path = os.path.join(root,file)
                new_path = os.path.join(root,new_file_name)
                if not os.path.exists(new_path):
                    os.rename(old_path,new_path)
                test_count+=1
```
